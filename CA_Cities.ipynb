```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

# htm from "http://www.beachcalifornia.com/city_county.html"
soup = BeautifulSoup(open("city_list.htm"))
```

```python
main_list = str(soup.find_all('tr')).split("\\n") # body str that contains city names
df_list = pd.Series(main_list)  # transform to Series
rv_item = [("href" not in main_list[i]) for i in xrange(len(main_list))]

city_list = df_list[rv_item][3:368]
city_list.index = xrange(len(city_list))
city_list = [item.split(",")[0].split(" ") for item in city_list]

def clean_name(str_cities):
    '''
    Takes a line of str from the city_list and clean up the format into a string of a city name
    '''
    # remove numbers
    np_cities = np.array(str_cities)
    rm_flag = np.array([any(char.isdigit() for char in np_cities[i]) for i in xrange(len(np_cities))])
    result = str(np_cities[~rm_flag])
    
    # remove "<br/>"
    result = result.replace("<br/>", "").replace("</a>", "").replace("'", "").replace("[", "").replace("]", "")
    
    # remove nonsence names
    if (len(result) > 50 or len(result) == 0 or "font" in result):
        result = ""
        
    return result

cities = [clean_name(line) for line in city_list]
city_names = filter(None, cities)  # 360 cities returned
```
